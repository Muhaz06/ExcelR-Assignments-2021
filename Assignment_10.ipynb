# Recommendation Systems

```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data = pd.read_csv('book.csv')
data
```

```python
data2 = data.iloc[:,1:]
data2
```

```python
data2.sort_values(['User.ID'])
```

```python
# number of unique users in the dataset
len(data2['User.ID'].unique())
```

```python
# number of unique books in the dataset
len(data2['Book.Title'].unique())
```

```python
# converting long data into wide data using pivot table
data3=data2.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
data3
```

```python
# Replacing the index values by unique user Ids
data3.index=data2['User.ID'].unique()
data3
```

```python
# Impute those NaNs with 0 values
data3.fillna(0,inplace=True)
data3
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(data3.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# Set the index and column names to user ids 
user_sim2.index=data2['User.ID'].unique()
user_sim2.columns=data2['User.ID'].unique()
user_sim2
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
# Most Similar Users
user_sim2.idxmax(axis=1)
```
