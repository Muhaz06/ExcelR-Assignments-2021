# Logistic Regression Model

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('bank-full.csv')
df
```

## EDA

```python
df.info()
```

```python
# One-Hot Encoding of categrical variables
# get_dummies() is used for data manipulation. It converts categorical data into dummy or indicater variables 
df2 =pd.get_dummies(df,columns=['job','marital','education','contact','poutcome','month'])
df2
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
df2
```

```python
df2.info()
```

```python
# Custom Binary Encoding of Binary o/p variables 
df2['default'] = np.where(df2['default'].str.contains("yes"), 1, 0)
df2['housing'] = np.where(df2['housing'].str.contains("yes"), 1, 0)
df2['loan'] = np.where(df2['loan'].str.contains("yes"), 1, 0)
df2['y'] = np.where(df2['y'].str.contains("yes"), 1, 0)
df2
```

```python
df2.info()
```

## Model Building 

```python
# Dividing our data into input and output variables
x=pd.concat([df2.iloc[:,0:10],df2.iloc[:,11:]],axis=1)
y=df2.iloc[:,10]
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

## Model Predictions

```python
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

## Testing Model Accuracy 

```python
# Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39155+1162)/(39155+767+4127+1162)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```

```python
# ROC Curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(y, classifier.predict_proba (x)[:,1])

auc = roc_auc_score(y, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```
